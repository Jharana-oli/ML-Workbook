```python
from bs4 import BeautifulSoup
print("BeautifulSoup is installed!")
```

```python
page_count=1
saved_pages=[]
while True:
    url=f"https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(url)
    soup=BeautifulSoup(res.text,"lxml")
    quote=soup.select("div.quote")
    if not quote:
        print("No valid pages anymore")
        break
    with open(f"scraped_data/page{page_count}.html","w", encoding="utf-8") as f:
        f.write(res.text)
        saved_pages.append(page_count)
        print(f"Downloaded data from page{page_count}")
    page_count+=1    
```

```python
life_quotes = []
for page in saved_pages:
    with open(f"scraped_data/page{page}.html", "r", encoding="utf-8") as f:
        html_content = f.read()
    soup = BeautifulSoup(html_content, "lxml")
    quotes = soup.select("div.quote")
    for q in quotes:
        tags = [tag.get_text() for tag in q.select(".tag")]
        if "life" in tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text, author])
print(life_quotes)
```
