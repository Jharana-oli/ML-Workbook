```python
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
cities = ["New York", "London", "Delhi", "Tokyo"]
temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]
```

```python
country_count=0
fig,axes=plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        axes[i][j].plot(days,temperatures[country_count])
        axes[i][j].set_title(cities[country_count])
        axes[i][j].grid(True)
        country_count+=1
fig.suptitle("Temperature in city over a week")      
fig.supxlabel("Days")
fig.supylabel("Temperature")
fig.tight_layout()
```
